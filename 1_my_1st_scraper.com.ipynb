## 2) Exploring HTML Structure with BeautifulSoup

### 2.1) Concepts

#### ðŸ§± What is the DOM?
DOM (Document Object Model) is a structured representation of a web page's HTML content, in the form of a hierarchical tree of nodes.

#### ðŸ§  What is it used for?
The DOM allows programming languages like JavaScript or Python (with BeautifulSoup, for example) to:
* Access HTML elements (`<div>`, `<h1>`, `<p>`, etc.)
* Read or modify content
* Navigate between elements (parents, children, siblings)
* Automate data extraction or manipulation

#### ðŸŒ³ What does the DOM look like?
For example, this HTML:

````python
# HTML Example
html_example = """
<html>
  <body>
    <h1>Web Scraping Course</h1>
    <p>Learn to extract web data</p>
  </body>
</html>
"""

print(html_example)
````

Is represented in the DOM as this tree:
```
html
â””â”€â”€ body
    â”œâ”€â”€ h1 â†’ "Web Scraping Course"
    â””â”€â”€ p â†’ "Learn to extract web data"
```

Each tag is a node, and can contain text or other nodes.

````python
from bs4 import BeautifulSoup

# First, define the HTML example
html_example = """
<html>
  <body>
    <h1>Web Scraping Course</h1>
    <p>Learn to extract web data</p>
  </body>
</html>
"""

# Parse the HTML
soup = BeautifulSoup(html_example, 'html.parser')

# Navigate through the DOM
title = soup.find('h1')
paragraph = soup.find('p')

print(f"Title: {title.text}")
print(f"Paragraph: {paragraph.text}")
````

## 3) Complete Product Extraction: Name, Price, and Image

We'll scrape book data from http://books.toscrape.com/

````python
# Define the URL
url = "http://books.toscrape.com/"

# Make the GET request
response = requests.get(url)

print(f"Status Code: {response.status_code}")
````

````python
# Parse the HTML content
soup = BeautifulSoup(response.text, "html.parser")

print("HTML parsed successfully!")
````

````python
# Find all product articles
products = soup.select("article.product_pod")

print(products)
````

````python
# Find all product articles
products = soup.select("article.product_pod")

print(f"Found {len(products)} products")
````

````python
# List to store the information
product_list = []

for product in products:
    # Book name
    nombre = product.find("h3").find("a")["title"]
    
    # Price
    precio = product.find("p", class_="price_color").get_text()
    
    # Image
    imagen = product.find("div", class_="image_container").find("img")["src"]
    imagen_url = "http://books.toscrape.com/" + imagen

    # Estrellas
    estrellas =  product.find("p", class_="star-rating")["class"][1]

    # Stock
    stock = product.find("p", class_="instock availability").get_text(strip=True)


    product_list.append({
        "name": nombre,
        "price": precio,
        "image_url": imagen_url,
        "stars": estrellas,
        "stock": stock
    })

# Display first 3 products as preview
print("First 5 products:")
for i, product in enumerate(product_list[:3], 1):
    print(f"\n{i}. {product['name']}")
    print(f"   Price: {product['price']}")
    print(f"   Image: {product['image_url']}")
    print(f"   Stars: {product['stars']}")
    print(f"   Stock: {product['stock']}")
````

````python
# Create the results directory if it doesn't exist
import os
os.makedirs("resultados", exist_ok=True)

# Save the data to CSV
path_csv = "resultados/productos.csv"

with open(path_csv, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "price", "image_url", "stars", "stock"])
    writer.writeheader()
    writer.writerows(product_list)

print(f"âœ… Extraction complete: {len(product_list)} products saved to {path_csv}")
````
